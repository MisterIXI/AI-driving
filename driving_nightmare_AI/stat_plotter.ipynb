# Flappy Bird

```python
dir = "fb_model"
stat_path = os.path.join(os.getcwd(), dir, "stats.pickle")
stats = []
if os.path.exists(stat_path):
    with open(stat_path, "rb") as f:
        stats = pkl.load(f)

scores = []
epsilon = []
loss = []
for stat in stats:
    scores.append(stat["score"])
    epsilon.append(stat["epsilon"])
    loss.append(stat["avg_loss"])
# n = len(scores)
n = 600
x = range(1, n+1)
scores = scores[:n]
epsilon = epsilon[:n]
loss = loss[:n]
fig, ax3 = plt.subplots()

color = 'tab:blue'
ax3.set_ylabel('Loss', color=color)
ax3.plot(x, loss, color=color,zorder=-1)
ax3.tick_params(axis='y', labelcolor=color)

ax1 = ax3.twinx()
color = 'tab:green'
ax1.set_xlabel('n')
ax1.set_ylabel('Time', color=color)
ax1.plot(x, scores, color=color, zorder=4)
ax1.tick_params(axis='y', labelcolor=color)

# ax2 = ax3.twinx()  
# color = 'tab:red'
# ax2.set_ylabel('Epsilon', color=color)  
# ax2.plot(x, epsilon, color=color)
# ax2.tick_params(axis='y', labelcolor=color)

score_series = pd.Series(scores)
moving_avg = score_series.rolling(window=70).mean()
ax1.plot(x, moving_avg, color='tab:orange', zorder=5)
ax1.legend(['Time', 'Moving Average', 'Loss'], loc='upper left')

# add label "Flappy bird"
fig.suptitle("Flappy Bird")
fig.tight_layout()  
plt.show()
```

# Driving Nightmare

```python
dir = "rl_data"
stat_path = os.path.join(os.getcwd(), dir, "stats.pickle")
stats = []
if os.path.exists(stat_path):
    with open(stat_path, "rb") as f:
        stats = pkl.load(f)

scores = []
epsilon = []
loss = []
for stat in stats:
    scores.append(stat["time"])
    epsilon.append(stat["epsilons"])
    loss.append(stat["avg_loss"])
n = len(scores)
x = range(1, n+1)

fig, ax1 = plt.subplots()

color = 'tab:green'
ax1.set_xlabel('n')
ax1.set_ylabel('Time', color=color)
ax1.plot(x, scores, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  
color = 'tab:red'
ax2.set_ylabel('Epsilon', color=color)  
ax2.plot(x, epsilon, color=color)
ax2.tick_params(axis='y', labelcolor=color)

ax3 = ax1.twinx()
color = 'tab:blue'
ax3.set_ylabel('Loss', color=color)
ax3.plot(x, loss, color=color)
ax3.tick_params(axis='y', labelcolor=color)

score_series = pd.Series(scores)
moving_avg = score_series.rolling(window=40).mean()
ax1.plot(x, moving_avg, color='tab:orange', zorder=5)
ax1.legend(['Time', 'Moving Average'], loc='upper left')

# add label "Flappy bird"
fig.suptitle("Driving Nightmare")
fig.tight_layout()  
plt.show()
```
